### Rueda omnidireccional

<div style="text-align: justify"> En los modelos cinemáticos anteriores se usaban ruedas simples, donde la rueda
sólo podia moverse (rodar) en la dirección de su orientación. Para permitir el movimiento omnidireccional (en
más direcciones) se requiere una construcción de rueda compleja. Un ejemplo de tal rueda es la rueda Mecanum (ver Fig.1) o rueda sueca donde el número de rodillos más pequeños están dispuestos alrededor de la llanta de la rueda principal </div> 

<!-- ![Alt Text](image path "title") -->
![Alt Text](1.png "Fig.1")

<div style="text-align: justify">Los ejes de los rodillos pasivos no están alineados con el eje de la rueda principal (normalmente
están en un $\gamma=\pi/4$ ). Esto permite una serie de movimientos en diferentes
direcciones, que resultan de una combinación arbitraria de la rueda principal y rotación de los rodillos pasivos.</div> 

### Cinemática de tracción omnidireccional en las cuatro ruedas

<div style="text-align: justify">La popular plataforma de cuatro ruedas conocida como plataforma Mecanum es
mostrado en la Fig.2. La plataforma Mecanum tiene ruedas con Rodillos orientados hacia la izquierda y derecha, donde las ruedas principales siempre pressentan la misma orientacion. Esto permite
vehículo  moverse en cualquier dirección con rotación arbitraria, ordenando
las velocidades adecuadas para cada rueda principal.</div> 

<!-- ![Alt Text](image path "title") -->
![Alt Text](2.png "Fig.2")

<div style="text-align: justify">El movimiento hacia adelante o hacia atrás se obtiene poniendo todas las ruedas principales a la misma velocidad. Si las velocidades de
las ruedas principales en un lado son opuestas a las velocidades de las ruedas en
el otro lado entonces la plataforma girará.</div> 

<div style="text-align: justify">El movimiento lateral de la plataforma se obtiene si las ruedas en una diagonal tienen velocidad opuesta a las ruedas
en la otra diagonal. Con la combinación de los movimientos descritos, el movimiento
de la plataforma en cualquier dirección y con cualquier rotación se puede lograr.</div> 

<div style="text-align: justify">La cinemática interna inversa del plataforma Mecanum de la Fig 3, se puede derivar de la siguiente manera. Las coordenadas se obtienen de la velocidad de la rueda principal $v_1(t)$ y la velocidad del rodillo $v_R(t)$.</div> 

<!-- ![Alt Text](image path "title") -->
![Alt Text](3.png "Fig.3")

<div style="text-align: justify">La velocidad total de la rueda en la dirección $x_m$ y $y_m$ del marco de coordenadas del robot es .</div> 
$$
v_{m1x}=v_1+v_R\cos(\frac{\pi}{4})=v_1+\frac{v_R}{\sqrt2}
$$

$$
v_{m1y}=v_R\sin(\frac{\pi}{4})=\frac{v_R}{\sqrt2}
$$

De lo cual se puede concluir que:

$$v_1=v_{m1x}-v_{m1y}$$

<div style="text-align: justify">La Velocidad de la primera rueda en la dirección del marco de coordenadas del robot puede también ser expresado de la siguiente manera:</div>

$$v_{m1x}=\dot{x}_{m}-\omega d$$

$$v_{m1y}=\dot{y}_{m}+\omega l$$

<div style="text-align: justify">Con las  ultimas relaciones la velocidad de la primera rueda puede ser expresada en la cuerpo del robot como:</div>

\begin{equation*}
v_1=\dot{x}_{m}-\omega d-\dot{y}_{m}-\omega l=\dot{x}_{m}-\dot{y}_{m}-(\omega d+\omega l)
\label{eq:1} \tag{1}
\end{equation*}

<div style="text-align: justify">Ahora se realiza lo mismo con las siguientes ruedas obteniendo. Para la rueda 2 tenemos:</div>

$$
v_{m2x}=v_2+v_R\cos(\frac{\pi}{4})=v_2+\frac{v_R}{\sqrt2}
$$

$$
v_{m2y}=-v_R\sin(\frac{\pi}{4})=-\frac{v_R}{\sqrt2}
$$

El signo negativo se debe a que $v_{m2y}$ esta hacia abajo

$$v_2=v_{m2x}+v_{m2y}$$

$$v_{m2x}=\dot{x}_{m}-\omega d$$

$$v_{m2y}=\dot{y}_{m}-\omega l$$

\begin{equation*}
v_2=\dot{x}_{m}-\omega d+\dot{y}_{m}-\omega l=\dot{x}_{m}+\dot{y}_{m}-(\omega d+\omega l)
\label{eq:2} \tag{2}
\end{equation*}
 Para la rueda 3 tenemos

$$
v_{m3x}=v_3+v_R\cos(\frac{\pi}{4})=v_2+\frac{v_R}{\sqrt2}
$$

$$
v_{m3y}=v_R\sin(\frac{\pi}{4})=\frac{v_R}{\sqrt2}
$$

$$v_3=v_{m3x}-v_{m3y}$$

$$v_{m3x}=\dot{x}_{m}+\omega d$$

$$v_{m3y}=\dot{y}_{m}-\omega l$$

\begin{equation*}
v_3=\dot{x}_{m}+\omega d-\dot{y}_{m}+\omega l=\dot{x}_{m}-\dot{y}_{m}+(\omega d+\omega l)
\label{eq:3} \tag{3}
\end{equation*}

 Para la rueda 4 tenemos

$$
v_{m4x}=v_4+v_R\cos(\frac{\pi}{4})=v_4+\frac{v_R}{\sqrt2}
$$

$$
v_{m4y}=-v_R\sin(\frac{\pi}{4})=-\frac{v_R}{\sqrt2}
$$

$$v_4=v_{m4x}+v_{m4y}$$

$$v_{m4x}=\dot{x}_{m}+\omega d$$

$$v_{m4y}=\dot{y}_{m}+\omega l$$

\begin{equation*}
v_4=\dot{x}_{m}+\omega d+\dot{y}_{m}+\omega l=\dot{x}_{m}+\dot{y}_{m}+(\omega d+\omega l)
\label{eq:4} \tag{4}
\end{equation*}

Las ecuaciones $\eqref{eq:1}$,$\eqref{eq:2}$,$\eqref{eq:3}$ y $\eqref{eq:4}$ pueden ser escritas en su forma matriciales resultando:

\begin{equation*}
\begin{bmatrix}
v_1 \\
v_2 \\
v_3 \\
v_4
\end{bmatrix}=
\begin{bmatrix}
1 & -1 & -(l+d)\\
1 & 1 & -(l+d)\\
1 & -1 & (l+d)\\
1 & 1 & (l+d)
\end{bmatrix}
\begin{bmatrix}
\dot{x}_{m}\\
\dot{y}_{m}\\
\omega
\end{bmatrix}
\label{eq:5} \tag{5}
\end{equation*}


<div style="text-align: justify">Ademas se debe tomar en cuenta que estas son las velocidades tangenciales de cada rueda por lo que se debe encontrar sus velocidades angulares.</div>
$$v_1=w_1 r$$

$$v_2=w_2 r$$

$$v_3=w_3 r$$

$$v_4=w_4 r$$

\begin{equation*}
\begin{bmatrix}
v_1 \\
v_2 \\
v_3 \\
v_4
\end{bmatrix}=
\begin{bmatrix}
r & 0 & 0 & 0 \\
0 & r & 0 & 0 \\
0 & 0 & r & 0 \\
0 & 0 & 0 & r \\
\end{bmatrix}
\begin{bmatrix}
w_1 \\
w_2 \\
w_3 \\
w_4 
\end{bmatrix}
\label{eq:6} \tag{6}
\end{equation*}

\begin{equation*}
\begin{bmatrix}
w_1 \\
w_2 \\
w_3 \\
w_4 
\end{bmatrix}=
\begin{bmatrix}
r & 0 & 0 & 0 \\
0 & r & 0 & 0 \\
0 & 0 & r & 0 \\
0 & 0 & 0 & r \\
\end{bmatrix}^{-1}
\begin{bmatrix}
1 & -1 & -(l+d)\\
1 & 1 & -(l+d)\\
1 & -1 & (l+d)\\
1 & 1 & (l+d)
\end{bmatrix}
\begin{bmatrix}
\dot{x}_{m}\\
\dot{y}_{m}\\
\omega
\end{bmatrix}
\label{eq:7} \tag{7}
\end{equation*}

\begin{equation*}
\begin{bmatrix}
w_1 \\
w_2 \\
w_3 \\
w_4 
\end{bmatrix}=
\begin{bmatrix}
1/r & -1/r & -(l+d)/r\\
1/r & 1/r & -(l+d)/r\\
1/r & -1/r & (l+d)/r\\
1/r & 1/r & (l+d)/r
\end{bmatrix}
\begin{bmatrix}
\dot{x}_{m}\\
\dot{y}_{m}\\
\omega
\end{bmatrix}
\label{eq:8} \tag{8}
\end{equation*}

### Matriz Transformacion Froma General

<div style="text-align: justify">Otra forma mas general de realizar este analisis es analizando de manera independiente cada sisteme de referencia del robot omnidireccional. Para lo cual se debe tomar en cuenta lo siguiente.</div>

<div style="text-align: justify">Necesitamos entender la cinemática de la rueda.
ilustrado en la Fig 4. En un marco $<X_w,Y_w>$ en el centro de la rueda. La velocidad lineal del centro de la rueda se escribe:</div>

<!-- ![Alt Text](image path "title") -->
![Alt Text](4.png "Fig.4")

$$
v_{xw}=v_{drive}-v_{slide}\sin(\gamma)
$$
$$
v_{yw}=v_{slide}\cos(\gamma)
$$

\begin{equation*}
\begin{bmatrix}
v_{xw} \\
v_{yw}
\end{bmatrix}=
\begin{bmatrix}
1 & -\sin(\gamma)\\
0 & \cos(\gamma)
\end{bmatrix}
\begin{bmatrix}
v_{drive}\\
v_{slide}
\end{bmatrix}
\label{eq:9} \tag{9}
\end{equation*}

$$
\mathbf{V_w}=\mathbf{L}\mathbf{V_d}
$$

\begin{equation*}
\begin{bmatrix}
w_{drive}
\end{bmatrix}=
\begin{bmatrix}
1/r & 0
\end{bmatrix}
\begin{bmatrix}
v_{drive}\\
v_{slide}
\end{bmatrix}
\label{eq:10} \tag{10}
\end{equation*}

$$
\mathbf{w}=\mathbf{N}\mathbf{V_d}
$$

<div style="text-align: justify">A continuacion se necesita incluir una tranformacion la cual sera utilizada si la rueda esta rotada  un cierto angulo generando un sistema de referencia $<X_b,Y_b> $.</div>

\begin{equation*}
\begin{bmatrix}
v_{xb} \\
v_{yb}
\end{bmatrix}=
\begin{bmatrix}
\cos(\beta) & -\sin(\beta)\\
\sin(\beta) &  \cos(\beta)
\end{bmatrix}
\begin{bmatrix}
v_{xw}\\
v_{yw}
\end{bmatrix}
\label{eq:11} \tag{11}
\end{equation*}

$$
\mathbf{V_b}=\mathbf{B}\mathbf{V_w}
$$

<div style="text-align: justify">La siguiente  transformación produce la velocidad lineal en la rueda en $<X_b,Y_b> $ con respecto a las velocidades del sistema de referencia $<X_m,Y_m>$ el cual es del robot movil.</div>

\begin{equation*}
\begin{bmatrix}
v_{xb} \\
v_{yb}
\end{bmatrix}=
\begin{bmatrix}
1 & 0 & -y_i\\
0 & 1 &  x_i
\end{bmatrix}
\begin{bmatrix}
v_{xm}\\
v_{ym}\\
\omega
\end{bmatrix}
\label{eq:12} \tag{12}
\end{equation*}

$$
\mathbf{V_b}=\mathbf{T}\mathbf{V_m}
$$

<div style="text-align: justify">Finalmente despejando todas las matrices llegamos a la siguiente tranformacion:</div>

$$
\mathbf{w_i}=\mathbf{N}\mathbf{L}^{-1}\mathbf{B}^{-1}\mathbf{T}\mathbf{V_m}
$$

<div style="text-align: justify">El vector fila que se obtien es el siguiente, tomar en cuenta q este seria el resultado solo d euna rueda por lo que la matriz final seria una matriz la cual se produzca haciendo stacking  de todas las ruedas</div>
$$
\mathbf{w_i}=\mathbf{h_i}\mathbf{V_m}
$$

$$
\mathbf{h_i}=
\begin{bmatrix}
\frac{\cos(\beta_i+\gamma_i)}{r\cos(\gamma_i)}&\frac{\sin(\beta_i+\gamma_i)}{r\cos(\gamma_i)}&-\frac{y_i\cos(\beta_i+\gamma_i)-x_i\sin(\beta_i+\gamma_i)}{r\cos(\gamma_i)}
\end{bmatrix}
$$

$$
\begin{bmatrix}
\mathbf{w_1} \\
\mathbf{w_2} \\
\mathbf{w_3} \\
\mathbf{w_4} \\
\end{bmatrix}
=\begin{bmatrix}
\mathbf{h_1} \\
\mathbf{h_2} \\
\mathbf{h_3} \\
\mathbf{h_4} \\
\end{bmatrix}\mathbf{V_m}
$$

##  Implementación del Sistema

### Importar las librerias necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
import time
import matplotlib as mpl
```

#### Tiempos para la simulacion

```python
t_sample=0.1
t_final=10+t_sample
t=np.arange(0,t_final,t_sample)
t=t.reshape(1,t.shape[0])
```

### Parametros del robot omnidireccional 

```python
a=0.15                         # se saco de las dimensiones reales del robot
b=0.15                           # se como no se mueve en b es cero
c=0.2355                       # se saco de las dimensiones reales l
d=0.1880                       # se saco de las diensiones reales w
r=0.05                         # radio de las ruedas
```

### Cinematica 
Se inicializa los vectores vacios para que sean almacenados en cada instante de tiempo

```python
xp=np.zeros((t.shape[0],t.shape[1]))
yp=np.zeros((t.shape[0],t.shape[1]))
```

### Velocidades Generales
Se inicializa los vectores vacios para que sean almacenados en cada instante de tiempo

```python
ul=0.5*np.ones((t.shape[0],t.shape[1]))
um=0.5*np.ones((t.shape[0],t.shape[1]))
w=0*np.ones((t.shape[0],t.shape[1]))
```

### Velocidades Angulares cada Rueda
Se inicializa los vectores vacios para que sean almacenados en cada instante de tiempo

```python
w_1=0*np.ones((t.shape[0],t.shape[1]))
w_2=0*np.ones((t.shape[0],t.shape[1]))
w_3=0*np.ones((t.shape[0],t.shape[1]))
w_4=0*np.ones((t.shape[0],t.shape[1]))
```

### Integracion
Se iniciliza los valores para que sean almacenados en cada instante de tiempo

```python
x=np.zeros((t.shape[0],t.shape[1]+1))
y=np.zeros((t.shape[0],t.shape[1]+1))
phi=np.zeros((t.shape[0],t.shape[1]+1))
```

### Condiciones iniciales del sistema

```python
phi[0,0]=(0*np.pi)/180
x_real=0
y_real=0
x[0,0]=x_real+a*np.cos(phi[0,0])-b*np.sin(phi[0,0])
y[0,0]=y_real+a*np.sin(phi[0,0])+b*np.cos(phi[0,0])
```

### Funcion del modelo del robot
En esta seccion se aplica la formula (16)

```python
def Jacobiano(v,phi,a,b):
    J=np.matrix([[np.cos(phi),-np.sin(phi),-(a*np.sin(phi)+b*np.cos(phi))],[np.sin(phi),np.cos(phi),a*np.cos(phi)-b*np.sin(phi)]])
    hp=J*v
    return hp
```

### Integracion numerica del sistema

```python
def euler(z,zp,t_sample):
    z=z+zp*t_sample
    return z
```

### Conversion de velocidades Generales a cada rueda
En esta seccion se utiliza la formula (13) 

```python
def conversion(v,c,d,r):
    T=np.array([[1,1,c+d],[1,-1,-c-d],[1,-1,c+d],[1,1,-c-d]])*(1/r)
    tranformacion_ruedas=T@v
    return tranformacion_ruedas[0,0],tranformacion_ruedas[1,0],tranformacion_ruedas[2,0],tranformacion_ruedas[3,0]
```

### Bucle simulacion

```python
for k in range(0,t.shape[1]):
    v=np.array([[ul[0,k]],[um[0,k]],[w[0,k]]])

    # Conversion a las velociades de las ruedas
    
    w_1[0,k],w_2[0,k],w_3[0,k],w_4[0,k]=conversion(v,c,d,r)
    
    ## Velocidad del punto modelado
    hp=Jacobiano(v,phi[0,k],a,b)
    
    ## Obtener las velocidades en Cada eje
    xp[0,k]=hp[0,0]
    yp[0,k]=hp[1,0]
    
    ## Integracion numerica
    x[0,k+1]=euler(x[0,k],xp[0,k],t_sample)
    y[0,k+1]=euler(y[0,k],yp[0,k],t_sample)
    phi[0,k+1]=euler(phi[0,k],w[0,k],t_sample)
```

### Funcion para las graficas del sistema

```python
def grafica(sty,titulo,x,y,etiqueta,ejex,ejey,color):
    mpl.style.use(sty)
    fig, ax = plt.subplots()
    ax.set_title(titulo.format(sty), color='0')
    ax.set_xlabel(ejex)
    ax.set_ylabel(ejey)
    ax.plot(x, y, color, label=etiqueta)
    ax.grid(linestyle='--', linewidth='0.2', color='black')
    legend = ax.legend(loc='upper right', shadow=False, fontsize='small')
    
```
